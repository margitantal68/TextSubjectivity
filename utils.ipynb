# Model performance vs number of epochs

```python
import pandas as pd

title = 'bert-finetuned model performance vs training epochs'

filename = 'results/bert-finetuned-epochs-SUBJ.csv'

df = pd.read_csv(filename)
df.shape
```

```python
from matplotlib import pyplot as plt

plt.plot(df['epochs'], df['accuracy'])
plt.plot(df['epochs'], df['F1'])
plt.legend(['Accuracy', 'F1'])
plt.title(title)
plt.xlabel('epochs')
plt.ylabel('Accuracy/F1')
plt.show()
```

# Model performance vs amount of training data

```python
import pandas as pd

# title = 'bert-finetuned model using limited training data'
title = 'embeddingbag model using limited training data'

# filename = 'results/bert_finetuned_limited_training_data.csv'
filename = 'results/embeddingbag_bert_tokenizer_limited_training_data.csv'

df = pd.read_csv(filename)
df.shape
```

```python
from matplotlib import pyplot as plt

plt.plot(df['percent']*100, df['accuracy'])
# plt.plot(df['percent']*100, df['precision'])
# plt.plot(df['percent']*100, df['recall'])
plt.plot(df['percent']*100, df['F1'])
plt.legend(['Accuracy', 'F1'])
# plt.legend(['Accuracy', 'F1'])
# plt.legend(loc="upper left")
ax = plt.gca()
# ax.set_xlim([xmin, xmax])
ax.set_ylim([0.85, 0.98])
plt.title(title)
plt.xlabel('Percentage of training data')
plt.ylabel('Accuracy/F1')
plt.show()

```
